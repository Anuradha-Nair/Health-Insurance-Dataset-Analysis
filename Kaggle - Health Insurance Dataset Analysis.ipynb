### üìä Step 2: Exploratory Data Analysis (EDA)


#### Steps:

    1. Summary statistics

    2. Distribution plots (age, BMI, charges)

    3. Categorical value counts (sex, smoker, region)

    4. Boxplots and violin plots to explore relationships

    5. Correlation heatmap

    6. Business insights and potential decisions

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual theme
sns.set(style="whitegrid")

# Summary statistics
summary_stats = df.describe(include='all')

# Plot distributions
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Age distribution
sns.histplot(df['age'], bins=20, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Age Distribution')

# BMI distribution
sns.histplot(df['bmi'], bins=20, kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('BMI Distribution')

# Charges distribution
sns.histplot(df['charges'], bins=30, kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Insurance Charges Distribution')

# Children count
sns.countplot(x='children', data=df, ax=axes[1, 1], palette='pastel')
axes[1, 1].set_title('Number of Children')

plt.tight_layout()
plt.show()

summary_stats
```

### üß† Insights from Summary Statistics & Distributions


#### 1. Age:

        Uniform distribution across 18 to 64.

        Median age is 39.

        --> Business Insight: Product offerings can be segmented by age bands (e.g., <30, 30‚Äì50, >50) for personalized insurance products.

#### 2. BMI: 

        Mean BMI ‚âà 30.6 ‚Äî borderline obese.

        Right-skewed distribution: some outliers above 45.

        --> Actionable Decision: Introduce wellness programs or discounts for customers maintaining healthy BMI levels.

#### 3. Charges:

        Highly right-skewed: mean ‚âà $13,270, but max ‚âà $63,770.

        Median is much lower ($9,382), indicating a few high-cost individuals.

        --> Business Action: Consider reinsurance or premium tiers for high-risk groups (e.g., older smokers with high BMI).

#### 4. Children
        Most customers have ‚â§ 2 children.

        --> Strategy: Family plans can be tiered based on number of children.

##### NEXT: Categorical value analysis and relationship plots like boxplots and heatmaps.

```python
# Count plots for categorical variables
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Sex
sns.countplot(x='sex', data=df, ax=axes[0], palette='coolwarm')
axes[0].set_title('Gender Distribution')

# Smoker
sns.countplot(x='smoker', data=df, ax=axes[1], palette='Set2')
axes[1].set_title('Smoker Distribution')

# Region
sns.countplot(x='region', data=df, ax=axes[2], palette='Set3')
axes[2].set_title('Region Distribution')

plt.tight_layout()
plt.show()
```

### üîç Categorical Analysis

#### ‚úÖ Gender Distribution:

        Balanced: ~50% male and female.

        --> No immediate business bias, but segmented marketing could still help.

#### üö¨ Smoker Status:

        Majority are non-smokers (~80%), but smokers may be driving high charges.

        --> Business Insight: Risk-adjusted pricing for smokers is essential.

#### üåç Region:

        Fairly balanced regional spread, with slightly more customers in the southeast.

        --> Targeted regional marketing campaigns or network partnerships could optimize operations.

#### NEXT: Exploring relationships between variables (e.g., smoker vs. charges, BMI vs. charges, etc.) using boxplots and heatmaps.

```python
# Relationship plots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Charges by smoker
sns.boxplot(x='smoker', y='charges', data=df, ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Charges by Smoker Status')

# Charges by region
sns.boxplot(x='region', y='charges', data=df, ax=axes[0, 1], palette='Set3')
axes[0, 1].set_title('Charges by Region')

# Charges by number of children
sns.boxplot(x='children', y='charges', data=df, ax=axes[1, 0], palette='coolwarm')
axes[1, 0].set_title('Charges by Number of Children')

# Charges vs BMI (with smoker hue)
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df, ax=axes[1, 1], palette='husl')
axes[1, 1].set_title('Charges vs BMI Colored by Smoker')

plt.tight_layout()
plt.show()
```

### üìà Relationship Insights

#### 1. Smoker Status vs. Charges

        Smokers pay significantly higher charges (often >$30,000).

        --> Business Action: High-risk group‚Äîconsider targeted intervention programs, higher premium brackets, or preventive care plans.

#### 2. Region vs. Charges

        No substantial difference in charges across regions.

        --> Action: Pricing strategy doesn't need regional customization, but regional campaigns may still help increase market share.

#### 3. Children vs. Charges

        No strong impact of number of children on charges.

        --> Action: Children might not heavily influence premiums ‚Äî could simplify family plan pricing tiers.

#### 4. BMI vs. Charges (Colored by Smoker)

        Smokers with high BMI are heavily concentrated at high charges.

        --> Action: These are high-risk individuals. Create targeted campaigns offering:

                * Smoking cessation support

                * Weight-loss incentives

                * Tiered pricing



### üî• Correlation Heatmap
